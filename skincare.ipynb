```python
print("üè∑Ô∏è Top 10 Brands by Product Count:")
print(df['product_name'].value_counts().head(10))
```

```python
print("üåü Top 10 Highest Rated Products:")
top_rated = df.sort_values(by='product_type', ascending=False).head(10)
print(top_rated[[ 'product_name', 'product_type', 'price']])
```

```python
print("üìä Bar Plot: Top 10 Brands by Product Count")
plt.figure(figsize=(10, 5))
df['product_name'].value_counts().head(10).plot(kind='bar', color='skyblue')
plt.title('Top 10 Brands by Product Count')
plt.xlabel('product_name')
plt.ylabel('Number of Products')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
print("üíµ Histogram: Product Price Distribution")
plt.figure(figsize=(8, 5))
sns.histplot(df['price'], bins=30, kde=True, color='lightgreen')
plt.title('Product Price Distribution')
plt.xlabel('Price')
plt.ylabel('Number of Products')
plt.tight_layout()
plt.show()
```
