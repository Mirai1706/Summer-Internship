```python
# Importing required libraries
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Reading the dataset
df = pd.read_csv('Webtoon Dataset.csv')

# Displaying the first 5 rows of the dataset
print("First 5 rows of the dataset:")
print(df.head())
```

```python
# Displaying shape of the dataset
print("Shape of the dataset (rows, columns):")
print(df.shape)

# Displaying data types and non-null values
print("\nInfo about the dataset:")
print(df.info())

# Checking for missing values
print("\nMissing values in each column:")
print(df.isnull().sum())
```

```python
# Displaying statistical summary
print("Statistical summary of numerical columns:")
print(df.describe())
```

```python
# Unique genres in the dataset
print("Unique genres:")
print(df['Genre'].unique())

# Count of webtoons in each genre
print("\nNumber of webtoons per genre:")
print(df['Genre'].value_counts())


```

```python
# Top 10 most viewed webtoons
print("Top 10 most viewed webtoons:")
print(df.sort_values(by='Likes', ascending=False).head(10))
```

```python
# Average rating for each genre
print("Average rating per genre:")
print(df.groupby('Genre')['Rating'].mean().sort_values(ascending=False))
```

```python


# Distribution of Ratings
print("Plotting distribution of Ratings...")
plt.figure(figsize=(8,5))
sns.histplot(df['Rating'], bins=20, kde=True)
plt.title('Distribution of Webtoon Ratings')
plt.xlabel('Rating')
plt.ylabel('Count')
plt.show()

# Top 10 genres with most webtoons
print("Plotting top 10 genres by count...")
plt.figure(figsize=(10,6))
df['Genre'].value_counts().head(10).plot(kind='bar', color='skyblue')
plt.title('Top 10 Genres with Most Webtoons')
plt.xlabel('Genre')
plt.ylabel('Count')
plt.xticks(rotation=45)
plt.show()
```

```python
# Pie chart showing percentage of webtoons in each genre
print("Pie chart: Distribution of Webtoons by Genre")
genre_counts = df['Genre'].value_counts()
plt.figure(figsize=(8,8))
plt.pie(genre_counts[:6], labels=genre_counts.index[:6], autopct='%1.1f%%', startangle=140)
plt.title('Top 6 Genres Distribution')
plt.axis('equal')  # Equal aspect ratio ensures pie is circular
plt.show()
```

```python
# Line plot for average ratings by genre
print("Line plot: Average Rating by Genre")
genre_ratings = df.groupby('Genre')['Rating'].mean().sort_values()
plt.figure(figsize=(12,6))
plt.plot(genre_ratings.index, genre_ratings.values, marker='o', linestyle='-')
plt.xticks(rotation=45)
plt.title('Average Rating by Genre')
plt.xlabel('Genre')
plt.ylabel('Average Rating')
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
# Box plot showing rating spread per genre
print("Box plot: Rating Distribution by Genre")
plt.figure(figsize=(14,6))
sns.boxplot(x='Genre', y='Rating', data=df)
plt.xticks(rotation=45)
plt.title('Rating Distribution by Genre')
plt.show()
```

```python
# Count of each status type
print("Bar plot: Webtoon Status Counts")
plt.figure(figsize=(6,4))
sns.countplot(x='Likes', data=df)
plt.title('Webtoon Status Count')
plt.xlabel('Likes')
plt.ylabel('Number of Webtoons')
plt.show()
```

```python
# Scatter plot for correlation between views and ratings
print("Scatter plot: Views vs Ratings")
plt.figure(figsize=(8,6))
sns.scatterplot(x='Likes', y='Rating', data=df)
plt.title('Likes vs Ratings')
plt.xlabel('Likes')
plt.ylabel('Rating')
plt.show()
```

```python
# Heatmap of numerical correlations
print("Heatmap: Correlation between numerical columns")
plt.figure(figsize=(8,6))
sns.heatmap(df.corr(numeric_only=True), annot=True, cmap='coolwarm', fmt='.2f')
plt.title('Correlation Heatmap')
plt.show()
```

```python
# Top 10 authors with most webtoons
print("Bar chart: Top 10 Authors by Number of Webtoons")
top_authors = df['Writer'].value_counts().head(10)
plt.figure(figsize=(10,5))
sns.barplot(x=top_authors.index, y=top_authors.values, palette='viridis')
plt.xticks(rotation=45)
plt.title('Top 10 Writer by Number of Webtoons')
plt.xlabel('Writer')
plt.ylabel('Number of Webtoons')
plt.show()
```
